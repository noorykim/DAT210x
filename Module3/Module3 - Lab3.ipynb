```python
fig = plt.figure()

ax = fig.add_subplot(111, projection='3d')
ax.set_xlabel('Area')
ax.set_ylabel('Perimeter')
ax.set_zlabel('Asymmetry')

ax.scatter(df.area, df.perimeter, df.asymmetry, c='red', marker='.')
plt.show()
```
